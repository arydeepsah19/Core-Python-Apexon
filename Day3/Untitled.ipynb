```python
#sort a list without using function and print only first 3 value in desending order

lis = [1,5,3,4,2,6]

for i in range(len(lis)):
    for j in range(i+1, len(lis)):
        if lis[i] > lis[j]:
            lis[i],lis[j] = lis[j], lis[i];
            
lis = lis[::-1]
            
print(lis[:3])
```

```python
# count the frequency of element with including special character

st = "Aryadeep $@Sah"
freq = {}

for ch in st:
    if('a'<=ch<='z') or ('A'<=ch<='Z'):
        if ch in freq:
            freq[ch]+=1
        else:
            freq[ch]=1
            
print(freq)
```

```python
#find the largest number in list

li = [1,33,53,5,11,4]

ma = li[0]

for i in range(len(li)):
    if li[i] > ma:
        ma = li[i]

print(ma)
    
```

```python
# st = I am a student. then it print like I ma a tneduts.

st = "I am a student."

word = st.split(" ")
new_word = []

for w in word:
    if w[-1].isalnum():
        new_word.append(w[::-1])
    else:
        new_word.append(w[:-1][::-1] + w[-1])

res = " ".join(new_word)

print(res)
```

```python
# find second largest in list

li = [1,33,53,5,11,4]

first = li[0]
second = li[0]

for i in range(len(li)):
    if li[i]>first:
        second = first
        first = li[i]
    elif li[i] > second and li[i]<first:
        second = li

print(second)
```

```python
#  first non-repeating character

st = "aabbcdeff"

freq = {}
res = ''

for ch in st:
    freq[ch] = freq.get(ch,0)+1
    
for ele in  freq:
    if freq[ele]==1:
        res = res+ele
        break
    
    
print(res)
```
