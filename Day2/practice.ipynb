```python
# Reverse a String

st = input("Enter original String : ")
print("Reversed String is : ",st[::-1])   # with the help of splice method


st1 ="".join(reversed(st))     # with the help of join and reversed method
print("Reversed String is : ",st1)
```

```python
# check palindrome

st = input("Enter the string : ")
st1 = "".join(reversed(st))

if(st==st1):
    print("String is palindrome")
else:
    print("String is not Palindrome")
```
